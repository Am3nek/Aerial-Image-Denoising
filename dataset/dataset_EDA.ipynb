```python
from PIL import Image
import os
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
import cv2
```

```python
noisy_image = "/Users/vedanshkumar/Documents/Fun_ml/Projects/FuzzyLogicFiltering/dataset/Dataset/Noisy/noisy_airport_2.jpg"
clean_image = "/Users/vedanshkumar/Documents/Fun_ml/Projects/FuzzyLogicFiltering/dataset/Dataset/Clean/airport_2.jpg"
```

```python
clean_img = cv2.imread(clean_image)
noisy_img = cv2.imread(noisy_image)
```

```python
plt.subplot(1,2,1)
plt.imshow(clean_img)
plt.subplot(1,2,2)
plt.imshow(noisy_img)
```

## Fourier Transform

```python
gray_clean = cv2.cvtColor(clean_img, cv2.COLOR_BGR2GRAY)
fourier_clean = cv2.dft(np.float32(gray_clean), flags=cv2.DFT_COMPLEX_OUTPUT)
clean_fourier_shift = np.fft.fftshift(fourier_clean)
magnitude_clean = 20*np.log(cv2.magnitude(clean_fourier_shift[:,:,0],clean_fourier_shift[:,:,1]))
plt.imshow(magnitude_clean)
```

```python
gray_noisy = cv2.cvtColor(noisy_img, cv2.COLOR_BGR2GRAY)
fourier_noisy = cv2.dft(np.float32(gray_noisy), flags=cv2.DFT_COMPLEX_OUTPUT)
noisy_fourier_shift = np.fft.fftshift(fourier_noisy)
magnitude_noisy = 20*np.log(cv2.magnitude(noisy_fourier_shift[:,:,0],noisy_fourier_shift[:,:,1]))
plt.imshow(magnitude_noisy)
```
